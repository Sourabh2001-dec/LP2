```python
import string
f = open("./data.txt",'r')
text = f.read()
data = "".join([char for char in text if char not in string.punctuation])
print(data)
```

```python
!pip download nltk
import nltk
nltk.download('stopwords')
from nltk.stem import PorterStemmer
word_stemmer = PorterStemmer()
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('tagsets')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
```

```python
nltk.download('omw-1.4')
```

```python
from nltk.tokenize import word_tokenize
tokened_sen = word_tokenize(data)
print(word_tokenize(data))
```

```python
nltk.pos_tag(tokened_sen)
```

```python
Filtered_sen = [w for w in tokened_sen if w not in stop_words]
print(Filtered_sen)
```

```python
for w in Filtered_sen:
  print(w + "---->"+word_stemmer.stem(w))
```

```python
for w in Filtered_sen:
  print(w + "---->"+lemmatizer.lemmatize(w))
```

```python
analyze = vectorizer.build_analyzer()
X = vectorizer.fit_transform(data.split('.'))
print(analyze(data))
```

```python
print('\nidf values:')
for ele1, ele2 in zip(vectorizer.get_feature_names(), vectorizer.idf_):
    print(ele1, ':', ele2)
```

```python
print('\nWord indexes:')
print(vectorizer.vocabulary_)
  
# display tf-idf values
print('\ntf-idf value:')
print(X)
  
# in matrix form
print('\ntf-idf values in matrix form:')
print(X.toarray())
```
