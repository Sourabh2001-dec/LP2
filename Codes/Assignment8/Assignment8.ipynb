```python
import pandas as pd
import numpy as np
import seaborn as sns
pd.set_option('mode.chained_assignment', None)
```

## Loading the inbuilt titanic dataset from seaborn

```python
titanic = sns.load_dataset('titanic')
```

```python
df = pd.DataFrame(titanic)
```

```python
df.head()
```

## Data cleaning

```python
df.isnull().sum()
```

```python
df.drop(["deck"],inplace=True,axis=1)
df["age"].fillna(df["age"].mean(),inplace=True)
df.embarked.fillna(df.embarked.mode()[0],inplace=True)
df.embark_town.fillna(df.embark_town.mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

## handling age outliers

```python
sns.boxplot(df["age"])
```

```python
Q1 = df["age"].quantile(0.25)
Q2 = df["age"].quantile(0.50)
Q3 = df["age"].quantile(0.75)

IQR = Q3 - Q1
low_limit = Q1 - (1.5*IQR)
up_lim = Q3 +( 1.5 * IQR)

df = df[(df["age"] >= low_limit) & (df["age"] <= up_lim)]

print(low_limit,up_lim)
```

```python
sns.boxplot(df["age"])
```

## handling outlier for fare

```python
sns.boxplot(df["fare"])
```

```python
Q1 = df["fare"].quantile(0.25)
Q2 = df["fare"].quantile(0.50)
Q3 = df["fare"].quantile(0.75)

IQR = Q3 - Q1
low_limit = Q1 - (1.5*IQR)
up_lim = Q3 +( 1.5 * IQR)

df = df[(df["fare"] > low_limit) & (df["fare"] < up_lim)]

print(low_limit,up_lim)
```

```python
sns.boxplot(df["fare"])
```

## Distplot

```python
sns.displot(df["fare"],kde=True)
```

## Jointplot

```python
sns.jointplot(x='fare',y='age',data=df)
```

## Rugplot

```python
sns.rugplot(data=df,x="fare",y="age")
```

```python
df_copy = df.copy()
cols = ["sex","embarked","class","who","adult_male","embark_town","alive","alone"]
df_copy.drop(cols,inplace=True,axis=1)
sns.pairplot(df_copy)
```
