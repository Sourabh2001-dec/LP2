### Importing required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("./iris.csv")
df = pd.DataFrame(dataset)
```

```python
df.head()
```

### Feature Names and data types

```python
df.dtypes
```

### Histogram to view all feature distributions

```python
noOfFeatures = len(df.columns)-1
```

```python
for i in range(noOfFeatures):
    plt.figure()
    sns.histplot(data=df,x=df.columns[i])
```

### Box plot of each feature to identify outliers

```python
for i in range(noOfFeatures):
    plt.figure()
    sns.boxplot(data=df,y=df.columns[i],x=df["variety"])
```
